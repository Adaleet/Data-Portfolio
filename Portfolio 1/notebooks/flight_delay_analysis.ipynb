```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
file_path = '/Users/adalatadiljan/Desktop/Portfolio 1/ot_delaycause1_DL/Airline_Delay_Cause.csv'
df = pd.read_csv(file_path)
print(df.info())
print(df.head())
print(df.columns)
```

```python
mean_arr_delay = df['arr_delay'].mean()
mean_carrier_delay = df['carrier_delay'].mean()
mean_weather_delay = df['weather_delay'].mean()
mean_nas_delay = df['nas_delay'].mean()
mean_security_delay = df['security_delay'].mean()
mean_late_aircraft_delay = df['late_aircraft_delay'].mean()
```

```python
print("Mean Arrival Delay:", mean_arr_delay)
print("Mean Carrier Delay:", mean_carrier_delay)
print("Mean Weather Delay:", mean_weather_delay)
print("Mean NAS Delay:", mean_nas_delay)
print("Mean Security Delay:", mean_security_delay)
print("Mean Late Aircraft Delay:", mean_late_aircraft_delay)
```

```python
delay_columns = ['arr_delay', 'carrier_delay', 'weather_delay', 'nas_delay', 'security_delay', 'late_aircraft_delay']
mean_values = [mean_arr_delay, mean_carrier_delay, mean_weather_delay, mean_nas_delay, mean_security_delay, mean_late_aircraft_delay]
```

```python
plt.bar(delay_columns, mean_values, color='skyblue')
plt.xlabel('Delay Type')
plt.ylabel('Mean Delay (minutes)')
plt.title('Mean Delays by Type')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
