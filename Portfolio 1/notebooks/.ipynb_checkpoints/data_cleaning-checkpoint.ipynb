```python
import pandas as pd

# Load the dataset
file_path = '/Users/adalatadiljan/Desktop/Portfolio 1/ot_delaycause1_DL/Airline_Delay_Cause.csv'
df = pd.read_csv(file_path)

# Display the first few rows of the dataframe
print("Initial Data:")
print(df.head())

# Check for missing values
print("\nMissing Values:")
print(df.isnull().sum())

# Drop columns that are not needed (replace 'column_to_drop' with actual column names)
columns_to_drop = ['column_to_drop1', 'column_to_drop2']  # Replace with actual columns
df.drop(columns=columns_to_drop, inplace=True, errors='ignore')

# Fill missing values (example: filling with mean for numeric columns)
for column in df.select_dtypes(include=['float64', 'int64']).columns:
    df[column].fillna(df[column].mean(), inplace=True)

# Alternatively, you can drop rows with missing values
# df.dropna(inplace=True)

# Convert data types (example: converting a column to datetime)
# df['date_column'] = pd.to_datetime(df['date_column'])  # Replace 'date_column' with actual column

# Check the data types after conversion
print("\nData Types After Cleaning:")
print(df.dtypes)

# Display the cleaned data
print("\nCleaned Data:")
print(df.head())

# Save the cleaned data to a new CSV file
cleaned_file_path = '/Users/adalatadiljan/Desktop/Portfolio 1/ot_delaycause1_DL/Airline_Delay_Cause.csv'  # Replace with your desired output file path
df.to_csv(cleaned_file_path, index=False)
print(f"\nCleaned data saved to: {cleaned_file_path}")
```
